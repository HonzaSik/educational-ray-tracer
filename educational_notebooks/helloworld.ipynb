## Hello world scene

```python
camera = PinholeCamera(
    origin=Vertex(0, 0, 4),
    direction=Vector(0, 0, -1),
    fov_deg=40.0,
)

light = PointLight(
    position=Vertex(5, 5, 5),
    intensity=1.0,
    color=Color.custom_rgb(255, 255, 255),
)

my_sphere = Object(
    geometry=Sphere(),
    material=PhongMaterial(base_color=Color.custom_rgb(255, 0, 0), spec_color=Color.custom_rgb(255, 255, 255), shininess=64, ior=1.0)
)

rt = LinearRenderLoop(
    scene=Scene(
        camera=camera,
        lights=[light],
        objects=[my_sphere],
    ),
)

img = rt.render("hello_world.png")
ipynb_display_images(img)
```
