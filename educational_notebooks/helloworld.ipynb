## Hello world scene

```python
camera = PinholeCamera(
    origin=Vertex(0, 0, 4),
    direction=Vector(0, 0, -1),
    fov_deg=40.0,
)

light = PointLight(
    position=Vertex(5, 5, 5),
    intensity=1.0,
    color=Color.custom_rgb(255, 255, 255),
)

noise = PerlinNoise(
    scale=5.0,
    strength=0.5,
)

my_sphere = Object(
    geometry=Sphere(),
    material=PhongMaterial(base_color=Color.custom_rgb(255, 0, 0), spec_color=Color.custom_rgb(255, 255, 255), shininess=64, transparency=0.95, ior=1.9, normal_noise=noise),
)

plane = Object(
    geometry=Square(vertex=Vertex(-5, -2, -5), diagonal_vertex=Vertex(5, -2, 5)),
    material=PhongMaterial(base_color=Color.custom_rgb(200, 200, 200), spec_color=Color.custom_rgb(255, 255, 255), shininess=16),
)

render_configuration = RenderConfig(
    resolution=Resolution.R720p,
    samples_per_pixel=15,
    max_depth=10,
)

rt = LinearRenderLoop(
    scene=Scene(
        camera=camera,
        lights=[light],
        objects=[my_sphere, plane],
        skybox="./skyboxes/sand.hdr"
    ),
    render_config=render_configuration,
    preview_config=PreviewConfig(
        progress_display=ProgressDisplay.TQDM_IMAGE_PREVIEW
    ),
)

img = rt.render("hello_world.png")
ipynb_display_images(img)
```
