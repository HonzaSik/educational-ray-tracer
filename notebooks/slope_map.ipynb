```python
from src.material.material.phong_material import *
from src import World
from src import Sphere
from src import PointLight
from src import Resolution
from src import Camera
from src import Vertex, Vector
from src import Scene
from src import ipynb_display_images
from src.render.loops import *
from src import Plane
from src.geometry.objects.cylinder import Cylinder

glossy_red = PhongMaterial(
    name="glossy_red",
    base_color=Color.custom_rgb(255, 0, 0),
    spec_color=Color.custom_rgb(255, 255, 255),
    shininess=50,
    ior=1.5,
    reflectivity=0.4,
    transparency=0.0
)

world = World()
world.add(
    Cylinder(
        base_point=Vertex(0, -1, -5),
        cap_point=Vertex(0, 1, -5),
        radius=1.0,
        material=glossy_red
    ),
)

point_light = PointLight(
    position=Vertex(5, 5, 0),
    intensity=1000.0,
    falloff=0.01
)

camera = Camera(
    fov=40,
    aspect_ratio=Resolution.R360p.aspect_ratio,
    origin=Vertex(0, 10, 3),
    direction=Vector(0, -1, -1),
)

scene = Scene(
    camera=camera,
    world=world,
    lights=[point_light],
    skybox_path=None
)

scene.validate()

my_ray_tracer = LinearRayCaster(scene=scene)
png_path = my_ray_tracer.render("images/my_first_render.png")
ipynb_display_images(png_path)
```
