# Loading the scene and rendering with the blinn-phong tracer at different resolutions

Now we can load the scene and render a preview at a chosen resolution (144p here for speed)

```python
scene = pm.load_scene("spheres")
```

```python
render_configuration = RenderConfig(
    resolution=Resolution.FullHD,
    samples_per_pixel=4,
    max_depth=6,
)

#todo fix TQDM_BAR not working because iam not updating the progress properly in the ray tracer
preview_configuration = PreviewConfig(
    progress_display=ProgressDisplay.TQDM_IMAGE_PREVIEW,
)

post_process_configuration = PostProcessConfig(
    enabled=True,
    scale_factor=2
)

my_ray_tracer = LinearRayCaster(scene=scene, render_config=render_configuration, preview_config=preview_configuration, post_process_config=post_process_configuration)
png = my_ray_tracer.render("phong_preview.png")

ipynb_display_image(png)
```
